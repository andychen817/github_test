# 1. Load Data

```python
from sklearn.datasets import load_wine
wine = load_wine()
```

```python
print(wine.DESCR)
```

Attribute Information:
- 1) Alcohol (酒精)
- 2) Malic acid (果酸)
- 3) Ash (灰值)
- 4) Alcalinity of ash (灰值的鹼度) 
- 5) Magnesium (鎂)
- 6) Total phenols (總酚)
- 7) Flavanoids (黃烷類)
- 8) Nonflavanoid phenols (Nonflavanoid酚)
- 9) Proanthocyanins (前花青素)
- 10)Color intensity (顏色強度)
- 11)Hue (顏色)
- 12)OD280/OD315 of diluted wines (OD280/OD315被稀釋的酒)
- 13)Proline (脯氨酸)


### 1-1. 轉成 pandas dataframe

```python
import pandas as pd
df_data = pd.DataFrame(data = wine.data, columns = wine.feature_names)
```

```python
df_data['target'] = wine.target
```

# 3. 訓練模型

### 3-1. 切分訓練/測試集

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_data, wine.target, test_size=0.2)
```

### 3-2. 選擇訓練模型

```python
from sklearn import ensemble 
model = ensemble.RandomForestClassifier()
```

### 3-3. model fit

```python
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

### 3-4. 評估模型準確率

```python
from sklearn import metrics
print("accuracy:%.2f"% metrics.accuracy_score(y_test, pred))
confusion_matrix = metrics.confusion_matrix(y_test, pred)
cm =pd.DataFrame(confusion_matrix , index = ['setosa', 'versicolor', 'virginica'], columns = ['setosa', 'versicolor', 'virginica'])
sns.heatmap(cm, annot = True)  
```
